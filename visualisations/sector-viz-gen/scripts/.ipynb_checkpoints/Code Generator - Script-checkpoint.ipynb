```python
import pandas as pd
from os import listdir
from os.path import isfile, join
```

```python
path = "../data/datasets/"
datasets = [f for f in listdir(path) if isfile(join(path, f))]
```

```python
def show_code(dataset):
    path = "../data/datasets/"
    df = pd.read_csv(path+dataset )
    
    print "\n" + dataset.split(".")[0]
    viz_path = "https://vizlab.openbudgetsindia.org/visualisations/sector-viz-gen/index.html#"
    for index, row in enumerate(df.iterrows()):
        print "\n" + row[1][1]
        print '<iframe width="100%" height="530px" src="' +viz_path + 'trend/barchart/' + dataset.split(".")[0] + '/'+ row[1][0]+'" frameborder="0"></iframe>'
    
    print "\n"
```

```python
for dataset in datasets:
    show_code(dataset)
```
